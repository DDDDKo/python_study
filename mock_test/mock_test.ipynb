<h3 style='font-weight: 700; color: #214151; background-color: #ffffff'>1. 함께 첨부된 officetel_estate.csv 파일을 DataFrame 형식의  officetel_estate 라는 변수로 불러오시오.</h3>

```python
# 1번 답안 
officetel_estae = pd.read_csv("officetel_estate.csv")
officetel_estae
```

<h3 style='font-weight: 700; color: #214151; background-color: #ffffff'>2. officetel_estate DataFrame 객체의 sequence 열을 행 인덱스로 변경하여 officetel_estate 변수에 저장하시오.</h3>

```python
# 2번 답안

officetel_estae = officetel_estae.set_index ('sequence')
officetel_estae
```

<h3 style='font-weight: 700; color: #214151; background-color: #ffffff'>3. officetel_estate DataFrame 객체에서 2021년의 행 만 출력하시오.</h3>

```python
# 3번 답안
officetel_estae[officetel_estae['year_month'].str.contains('2021')]
officetel_estae[officetel_estae.year_month.str.contains('2021')]
```

<h3 style='font-weight: 700; color: #214151; background-color: #ffffff'>4. officetel_estate DataFrame 객체에서 locale을 기준으로 그룹화하여 각 locale의 매매가 대비 전세가 (lease_to_sale_ratio)의 평균을 DataFrame으로 출력하시오. (평균은 소수점 이하 3번째 자리에서 반올림하시오.)</h3>

```python
# 4번 답안
officetel_estae.groupby(officetel_estae.locale).lease_to_sale_ratio.mean().round(2)         # 시리즈로 출력
officetel_estae.groupby(officetel_estae.locale)[['lease_to_sale_ratio']].mean().round(2)   # 데이터프레임으로 출력
```

<h3 style='font-weight: 700; color: #214151; background-color: #ffffff'>5. officetel_estate DataFrame 객체에서 행 인덱스로 year_month, 열 인덱스로 locale, 데이터로 average_sale_price 를 가지는 피봇 테이블을 출력하시오.</h3>

```python

# 5번 답안
officetel_estae.pivot_table('average_sale_price',  'year_month',  'locale')
officetel_estae.pivot_table(index='year_month', columns= 'locale', values='average_sale_price')
```

<h3 style='font-weight: 700; color: #214151; background-color: #ffffff'>6. officetel_estate DataFrame 객체에서 busan의 year_month (2022년)에 대한 average_sale_price 변화를 나타내는 선 그래프를 pandas의 plot을 활용하여 아래와 같이 출력하시오.</h3>

<img src='./mock_image/6.png' />

```python
# 6번 답안
# aaa = officetel_estae[officetel_estae['year_month'].str.contains('2021')]
# aaa.loc[officetel_estae['locale'] == 'busan', ['average_sale_price']].plot(index='year_month', columns='locale')
officetel_estae[officetel_estae['year_month'].str.contains('2021')].pivot(index='year_month', columns='locale', values='average_sale_price').busan.plot()
plt.title('Average sale price change in Busan in 2021')
plt.xlabel('Year - Month')
plt.ylabel('Average Sale Price')
plt.grid(True)
plt.show()
```

<h3 style='font-weight: 700; color: #214151; background-color: #ffffff'>7. officetel_estate DataFrame 객체에서 busan, ulsan, daegu의 year_month (2022년)에 대한 average_sale_price 변화를 나타내는 막대 그래프를 pandas의 plot을 활용하여 아래와 같이 출력하시오. (figure의 사이즈는 12, 8)</h3>

<img src='./mock_image/7.png' />

```python
# 7번 답안

officetel_estae[officetel_estae['year_month'].str.contains('2021')].pivot(index='year_month', columns='locale', values='average_sale_price')[['busan', 'ulsan', 'daegu']].plot.bar(figsize=(12, 6), rot=0)
plt.title('Average sale price change in Busan, Ulsan, Daegu in 2022')
plt.ylim([120000, 180000])
plt.xlabel('Year - Month')
plt.ylabel('Average Sale Price')
plt.grid(True)
plt.show()
```

<h3 style='font-weight: 700; color: #214151; background-color: #ffffff'>8. officetel_estate DataFrame 객체에서 locale을 기준으로 그룹화하여 각 locale의 매매가 대비 전세가 (lease_to_sale_ratio)의 평균 중 서울의 막대그래프를  pandas의 plot을 활용하여 아래와 같이 출력하시오. (figure의 사이즈는 12, 8, 평균은 소수점 이하 3번째 자리에서 반올림하시오.)</h3>

<img src='./mock_image/8.png' />

```python
officetel_estae[officetel_estae.locale.str.contains('seoul')].groupby(officetel_estae.locale)[['lease_to_sale_ratio']].mean().round(2)  # 데이터프레임으로 출력 => 막대도표가 무엇을 의미하는지 출력
officetel_estae[officetel_estae.locale.str.contains('seoul')].groupby(officetel_estae.locale).lease_to_sale_ratio.mean().round(2) # 시리즈로 출력 => 막대도표가 무엇을 의미하는지 출력 x
```

```python
# 8번 답안
officetel_estae[officetel_estae.locale.str.contains('seoul')].groupby(officetel_estae.locale).lease_to_sale_ratio.mean().round(2).plot.bar(figsize=(12, 6), rot=0)
plt.title('Average Lease To Sale Ratio by Locale')
plt.ylim([70, 90])
plt.xlabel('Locale')
plt.ylabel('Lease To Sale Ratio')
plt.grid(True)
plt.show()
```
