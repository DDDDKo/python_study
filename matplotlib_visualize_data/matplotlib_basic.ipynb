### 데이터 시각화 패키지 맷플롯리브
matplotlib 패키지는 파이썬에서 데이터를 차트 혹은 플롯으로 시각화 하는 패키지이다  
- 선형 그래프(line plot)
- 점 그래프(scatter plot)
- 등고선 (contour plot)
- 평면도 (surface plot)
- 막대 그래프 (bar chart)
- 히스토그램 (histogram)
- 박스 그래프 (box plot)

````python
# matplotlib 한글 표시 설정
import matplotlib.font_manager as fm

# matplotlib의 글꼴 관리자에 특정 글꼴을 추가
fe = fm.FontEntry(
    fname=r'font/NanumGothic.ttf',
    name='NanumGothic'
)
fm.fontManager.ttflist.insert(0, fe)

# pyplot에 matplotlib의 글꼴 관리자에 등록된 글꼴 적용
plt.rcParams.update({'font.family': 'NanumGothic', 'font.size': 14})
# pyplot에 `-`를 문자로 취급하도록 설정
plt.rcParams['axes.unicode_minus'] = False
````

### 차트의 종류
  
**막대 그래프** : 데이터의 카테고리별 값을 막대로 나타내는 방식 (수직 막대 그래프, 수평 막대 그래프)  
> 카테고리별로 비교할 때 주로 사용된다 (제품별 판매량, 지역별 인구수)  

**선 그래프** : 시간의 흐름에 따른 데이터 포인터를 선으로 연결하여 나타내는 방식  
> 시간의 흐름에 따라서 비교할 때 주로 사용된다 (기온의 변화, 주식 가격)  

**파이 차트** : 전체에 대한 비율을 원형으로 나타내는 방식 (각 조각의 크기가 해당 부분의 비율 표시)  
> 전체에 대한 비율을 비교할때 사용된다 (여론조사 결과)  

**산점도** : 두 변수 간의 상관 관계를 점들의 분포로 나타내는 방식  
> 두 변수의 성관 관계를 파알할 때 주로 사용된다 (키와 몸무게의 상관관계)  

**히스토그램** : 범위별로 그룹화해서 각 그룹에 속하는 데이터를 포인터의 수를 막대로 나타내는 방식
> 데이터의 분포를 비교할때 주로 사용된다 (시험 점수의 분포)  
  
**영역 그랲프** : 선 그래프의 아래 영역에 색으로 채운 형태
> 시간의 흐름에 따라 비교할 때 주로 사용된다, 변화량을 강조할 때 (누적 감염자, 누적 수출량)

**박스 그래프** : 데이터의 분포를 나타내는 방식, 최소값, 최대값, 사분위수를 나타내는 방식  
> 데이터의 이상치를 판단하여 주로 사용됨()

pyplot 하위 패키지의 `plot()` 메서드로 간단하게 생성 가능하다, 매개변수로 리스트 혹은 배열을 전달

### 차트 스타일 지정
'plot()' 메서드에 특정한 문자열 매개변수를 전달하면 스타일을 변경 할 수 있다  

````python
plt.title('rs-- x 틱')
plt.plot([2000, 2005, 2010, 2015, 2020, 2025], [1, 2, 4, 8, 16, 32], 'c1-.')
plt.show
````

````python
plt.title('스타일 지정')
plt.plot([2000, 2005, 2010, 2015, 2020, 2025], [1, 2, 4, 8, 16, 32], lw=5, c='#00998850', ls=':', marker='*', ms=10, mfc='#009988')
plt.show
````

### `plot()` 메서드의 스타일 배개변수
- color (c) : 선 색상  
- linewidth(lw) : 선의 굵기  
- linestyle(ls) : 선의 스타일  
- marker : 마커 종류  
- markersize (ms) : 마커 크기  
- markeredgecolor (mec) : 마커 테두리 색상  
- markeredgewidth (mew) : 마커 테두리 굵기  
- markerfacecolor (mfc) : 마커 내부 색상

### 차트 범위 지정
`xlim()`, `ylim()` 메서드로 x축과 y축의 상/하한 값을 지정할 수 있음

````python
plt.title('차트 범위 변경')
plt.plot([2000, 2005, 2010, 2015, 2020, 2025], [1, 2, 4, 8, 16, 32], 
         lw=5, c='#00998850', ls=':', marker='*', ms=10, mfc='#009988')
plt.xlim(1995, 2030)
plt.ylim(-1, 40)
plt.show
````

### 틱 설정
`xticks()`, `yticks()` 메서드로 틱의 위치나 라벨을 직접 지정할 수 있다

````python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title('틱 설정')
plt.plot(X, C, c='#00553090', ls=':')
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi]) # , ['아', '이', '우', '에', '오'])
plt.yticks([-1, 0, 1], ['Low', 'Zero', 'High'])
plt.show
````

### Grid 설정
`grid()` 메서드로 그리드를 표현할지 말지를 정할 수 있다

````python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title('틱 설정')
plt.plot(X, C, c='#00309060', ls=':')
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi]) # , ['아', '이', '우', '에', '오'])
plt.yticks([-1, 0, 1], ['Low', 'Zero', 'High'])
plt.grid(True)
plt.show
````

### 여러 그래프를 하나의 차트에 표현하기
`plot()` 메서드에 x, y, style 매개변수를 반복해서 전달하면 여러개의 그래프를 하나의 차트에 표현할 수 있다

````python
t = np.arange(0., 5., 0.2)
plt.title('여러 그래프 표현')
plt.plot(t, t, 'r--', t, 0.5*t**2, 'bs:', t, 0.2*t**3, 'g^-')
plt.grid(True)
plt.show
````

### 차트 겹치기
`plot()` 메서드를 중복해서 작성하여 차트를 겹칠 수 있음

````python
plt.title('차트 겹치기')
plt.plot([2000, 2005, 2010, 2015, 2020, 2025], [1, 2, 4, 8, 16, 32], 
         lw=5, c='#99880050', ls=':', marker='*', ms=10, mfc='#998800')
plt.plot([2000, 2005, 2010, 2015, 2020, 2025], [32, 16, 8, 4, 2, 1], 
         lw=5, c='#00998850', ls=':', marker='o', ms=10, mfc='#009988')
plt.show
````

### 범례
여러 개의 그래프가 한 차트에 표시 될 때 각 그래프가 무엇을 의미하는지 보여주는 범례를 표현할 수 있다  
`legend()` : 메서드로 표현(`plot()` 메서드에 `label` 매개변수를 지정해야 한다)  
위치를 수동으로 지정하고 싶으면 `loc()` 매개변수를 사용해야 한다

````python
plt.title('범례 표현')
plt.plot([2000, 2005, 2010, 2015, 2020, 2025], [1, 2, 4, 8, 16, 32], 
         lw=5, c='#99880050', ls=':', marker='*', ms=10, mfc='#998800', label='증가')
plt.plot([2000, 2005, 2010, 2015, 2020, 2025], [32, 16, 8, 4, 2, 1], 
         lw=5, c='#00998850', ls=':', marker='o', ms=10, mfc='#009988', label='감소')
plt.legend(loc=10)
plt.show
````

### x축, y축 라벨
`xlabel()`, `ylabel()` 메서드로 x축과 y축의 라벨을 표시할 수 있다

````python
plt.title('각 축 라벨 표현')
plt.plot([2000, 2005, 2010, 2015, 2020, 2025], [32, 16, 8, 4, 2, 1], 
         lw=5, c='#00998850', ls=':', marker='o', ms=10, mfc='#009988', label='감소')
plt.xlabel('년도')
plt.ylabel('증가치')
plt.show
````

### 차트 그림의 구조
matplotlib 으로 그려지는 차트는 Figure 객체, Axes 객체, Axis 객체로 구성이 되어져있다  
Figure 객체는 하나 이상의 Axes 객체를 포함하고있고, Axes 객체는 두 개 이상의 Axis 객체를 포함하고 있다  
  
- `Figure` : 차트가 그려지는 캔버스
- `Axes` : 하나의 플롯(차트, 그래프)
- `Axis` : 하나의 축

#### Figure 객체
모든 차트 그래프 결과는 Figure 객체에 포함되어 있다  
내부 차트가 아닌 경우 하나의 Figure 객체에는 하나의 아이디와 윈도우를 가진다  
일반적으로 기본 파이썬에서 차트를 표시하면 Figure 하나당 하나의 윈도우를 가진다  
`plot()` : 메서드가 자동으로 Figure 객체를 생성하기 때문에 직접 생성하는 경우는 드물지만   
　　　　여러개의 윈도우를 동시에 출력할 때나 차트 전체의 크기를 변경하고 싶을때 사용할 수 있다

````python
f1 = plt.figure(figsize=(10, 2))
plt.title('figure 객체 사이즈 변경')
plt.plot(np.random.randn(100))
plt.show
print(f1)
````

현재 사용하고있는 Figure 객체를 얻으려면 `gcf()` 메서드로 얻을 수 있다

````python
f1 = plt.figure(figsize=(10, 2))
plt.title('figure 객체 사이즈 변경')
plt.plot(np.random.randn(100))

f2 = plt.gcf()
print(id(f1))
print(id(f2))
plt.show

plt.title('figure 객체 사이즈 변경')
plt.plot(np.random.randn(100))

f3 = plt.gcf()
print(id(f3))
````
